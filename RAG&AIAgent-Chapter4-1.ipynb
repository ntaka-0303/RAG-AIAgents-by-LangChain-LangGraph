```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model_name="gpt-4o-mini", temperature=0)
messages = [
    SystemMessage("You are a helpful assistant"),
    HumanMessage("こんにちは。私は高松です"),
    AIMessage("こんにちは、高松さん！どのようにお手伝いできますか？"),
    HumanMessage("私の名前がわかりますか？"),
]

ai_message = model.invoke(messages)
print(ai_message.content)
```

```python
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model_name="gpt-4o-mini", temperature=0)
messages = [
    SystemMessage("You are a helpful assistant"),
    HumanMessage("こんにちは。私は高松です"),
]
chunks = []
full_message = None
for chunk in model.stream(messages):
    chunks.append(chunk)
    print(chunk.text)
    full_message = chunk if full_message is None else full_message + chunk
```

```python
from pydantic import BaseModel, Field
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

class Recipe(BaseModel):
    ingredients: list[str] = Field(description="材料リスト")
    steps: list[str] = Field(description="手順")

prompt = PromptTemplate.from_template("""以下の料理のレシピを教えてください。

料理名：{dish}""")

model = ChatOpenAI(model_name="gpt-4o-mini", temperature=0)

model_with_schema = model.with_structured_output(Recipe)

chain = prompt | model_with_schema

result = chain.invoke({"dish": "カレーライス"})

print(result)
print(result.model_dump_json(indent=2))
```

```python
from langchain_core.output_parsers import StrOutputParser

chain2 = prompt | model | StrOutputParser()

result = chain2.invoke({"dish": "カレーライス"})

print(result)
```
